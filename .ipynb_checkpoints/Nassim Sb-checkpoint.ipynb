```python
import pandas as pd
from pybliometrics.scopus import SubjectClassifications
from pybliometrics.scopus import AuthorRetrieval
import matplotlib.pyplot as plt
```

```python
new_query = {'description': 'Computer Science'}
sub = SubjectClassifications(new_query)
data = {'code': [], 'description': [], 'detail': [], 'abbrev': []}

for subject in sub.results:
    data['code'].append(subject.code)
    data['description'].append(subject.description)
    data['detail'].append(subject.detail)
    data['abbrev'].append(subject.abbrev)

df = pd.DataFrame(data)
```

```python
df.head(5)
```

```python
df['detail'].unique()
```

```python
query2 = {'description': 'Mathematics'}
sub2 = SubjectClassifications(query2)
data2 = {'code': [], 'description': [], 'detail': [], 'abbrev': []}
for subject in sub2.results:
    data2['code'].append(subject.code)
    data2['description'].append(subject.description)
    data2['detail'].append(subject.detail)
    data2['abbrev'].append(subject.abbrev)
```

```python
df2=pd.DataFrame(data2)
```

```python
df2.head(5)
```

```python
df2['detail'].unique()
```

```python
main = pd.concat([df, df2], ignore_index=True)
```

```python
main.head()
```

```python
main['detail'].unique()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(main['detail'], main['description'], test_size=0.2, random_state=42)
```

```python
vectorizer = TfidfVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train_vectorized, y_train)
```

```python
y_pred = model.predict(X_test_vectorized)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
print('Classification Report:\n', classification_report(y_test, y_pred))
```

```python
# Create a ML pipeline
```

```python
classifiers = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier()
}
```

```python
classifier_accuracy = {}
for name, clf in classifiers.items():
    clf.fit(X_train_vectorized, y_train)
    y_pred = clf.predict(X_test_vectorized)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Classifier: {name}")
    print(f"Accuracy score: {accuracy}")
    classifier_accuracy[name] = accuracy
   
```

```python
plt.figure(figsize=(8, 6))
plt.bar(classifier_accuracy.keys(), classifier_accuracy.values(), color='skyblue')
plt.xlabel('Classifiers')
plt.ylabel('Accuracy')
plt.title('Accuracy of Different Classifiers')
plt.ylim(0, 1) 
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```
